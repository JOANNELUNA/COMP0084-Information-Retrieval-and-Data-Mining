### Q1. Text Preprocessing Choice(s) and Size of Vocabulary:

For this task, I will extract the terms (1-grams) from the raw text in the passage-collection.txt file. I will perform the following preprocessing steps:

Open the input file specified by the filename parameter in read mode and set the encoding to UTF-8.

Initialize a WordNetLemmatizer object called lemmatizer, which is used to lemmatize each word in the text.

Tokenize each line into individual words using the word_tokenize() function from the nltk.tokenize module.

Apply lemmatization to each token using the lemmatize() method of the lemmatizer object.

Check if the resulting lemma is an alphabetic string using the isalpha() method. If it is, add it to the vocabulary set.

After processing all lines in the input file, return the size of the vocabulary set, which corresponds to the number of unique lemmatized words in the text.

In summary, the get_vocabulary_size() function performs the following preprocessing steps: tokenization and lemmatization. Stop word removal is not performed in the function, so all words (including stop words) will be added to the vocabulary set. The resulting vocabulary set contains the unique lemmas of all words in the text after these preprocessing steps have been applied.

Overall, these preprocessing methods help us to clean and prepare our text data for analysis by reducing noise, grouping related words together, and creating a more structured and organized representation of the text. And the size of the identified index of terms (vocabulary) is 98562 after these preprocessing steps have been applied.

```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
import numpy as np


nltk.download('punkt')
nltk.download('wordnet')
```

```python
def get_vocabulary_size(filename):
    vocabulary = set()
    lemmatizer = WordNetLemmatizer()
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            tokens = word_tokenize(line.lower())
            for token in tokens:
                lemma = lemmatizer.lemmatize(token)
                if lemma.isalpha():
                    vocabulary.add(lemma)
    return len(vocabulary)
    
get_vocabulary_size('passage-collection.txt')
```


### Q2. Term Occurrence and Zipf's Law:

I will now implement a function that counts the number of occurrences of terms in the provided dataset and plot their probability of occurrence (normalised frequency) against their frequency ranking. The function reads each line of the file, converts it to lowercase, and extracts each term using regular expressions. It then counts the number of occurrences of each term and returns a dictionary of term counts.

```python
def count_terms(filename):
    term_counts = {}
    lemmatizer = WordNetLemmatizer()
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            tokens = word_tokenize(line.lower())
            for token in tokens:
                lemma = lemmatizer.lemmatize(token)
                if lemma.isalpha():
                    term_counts[lemma] = term_counts.get(lemma, 0) + 1
    return term_counts

term_counts = count_terms('passage-collection.txt')
# term_counts
```

I will use this function to count the occurrences of terms in the file passage-collection.txt and plot the probability of occurrence against their frequency ranking on a log-log scale. This will help to visualize the distribution more clearly and compare it to the expected Zipfian distribution for text sets $s = 1$ defined by equation below. 

$$f(k; s, N) = \frac{k^{-s}}{\sum_{i=1}^{N} i^{-s}}$$

where $k$ is the frequency rank of a term (i.e., the position of the term in the frequency ranking), $s$ is a distribution parameter that determines the skewness of the distribution (i.e., the degree to which high-frequency terms are more common than low-frequency terms), $N$ is the total number of terms in the text corpus.

```python
def plot_zipf(term_counts):

    total_count = sum(term_counts.values())
    term_freq = [(term_counts[term] / total_count) for term in term_counts] # norm freq
    assert np.isclose(sum(term_freq),1), "frequency is not normalised"
    sorted_term_freq = sorted(term_freq, reverse=True)
    

    rank = range(1, len(sorted_term_freq) + 1)
    zipf_dist = [1/(i) for i in rank]

    plt.plot(rank, sorted_term_freq,label = 'Empirical')
    plt.plot(rank, zipf_dist, label = "Zipfian")
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Rank')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

plot_zipf(term_counts)
```

I observe that the empirical distribution has a lower frequency of high-ranking terms and a lower frequency of low-ranking terms compared to the Zipfian distribution and the slope of the empirical distribution is -1, it suggests that the text corpus without removing stop words does not follow Zipf's law very closely but still exhibits some degree of scaling.

The lower frequency in the empirical distribution compared to the Zipfian distribution suggests that this text corpus contains a higher proportion of medium-frequency terms that are not well modeled by Zipf's law. This could be due to factors such as the specific domain or topic of the corpus, the presence of stop words, or the preprocessing steps applied to the corpus.

However, if the slope of the log-log plot of the empirical distribution is close to -1, it indicates that the empirical distribution still exhibits a power-law relationship between term frequency and rank, albeit with deviations from Zipf's law. This slope is expected because the Zipfian distribution predicts a power-law relationship with a slope of -1, and the empirical distribution is expected to follow a similar relationship with some deviations due to the factors mentioned above.


### Q3. Quantitative Comparison between Empirical and Actual Zipf's Law Distributions:

To compare the two distributions more quantitatively, we can calculate the difference between the empirical and Zipfian frequencies at each rank and plot it on a log-log scale. This plot is called the log-difference plot and can be used to identify deviations from Zipf's law.

The difference between the empirical distribution and the idealized Zipfian distribution can be quantified by calculating the goodness-of-fit measure, which is the sum of the squared errors between the observed and expected frequencies at each rank. And the goodness-of-fit measure of 1.4297 suggests that there is a slight deviation between the empirical distribution of terms in the text corpus and the idealized Zipfian distribution.

```python
def Diff(term_counts):

    total_count = sum(term_counts.values())
    term_freq = [(term_counts[term] / total_count) for term in term_counts] # norm freq
    assert np.isclose(sum(term_freq),1), "frequency is not normalised"
    sorted_term_freq = sorted(term_freq, reverse=True)
    

    rank = range(1, len(sorted_term_freq) + 1)
    zipf_dist = [1/(i) for i in rank]

    log_diff = [abs(x - y) for x, y in zip(sorted_term_freq, zipf_dist)]
    
    # ratio = [sorted_term_freq[i] / zipf_dist[i] for i in range(len(rank))]
    squared_errors = [(observed - expected) ** 2 for observed, expected in zip(sorted_term_freq, zipf_dist)]
    gof_measure = sum(squared_errors)

    plt.plot(rank, log_diff)
    # plt.plot(rank, squared_errors)
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Rank')
    plt.ylabel('Diff')
    plt.show()
    return gof_measure
    
    
Diff(term_counts)
```

o explain the difference between the empirical distribution and the Zipfian distribution, we can consider the effect of the distribution parameter s. The parameter s determines the skewness of the distribution, with larger values of s producing more skewed distributions. In the case of Zipf's law (where s = 1), the distribution is relatively less skewed.

However, the actual frequency of a term is influenced by various factors such as syntax, grammar, and semantics, which can cause deviations from the ideal Zipfian distribution. Specifically, high-ranking terms tend to be more specific and less common than low-ranking terms. This means that the frequency of high-ranking terms in the empirical distribution is lower than predicted by the Zipfian distribution.

In summary, the difference between the empirical distribution and the Zipfian distribution can be explained by the fact that the actual frequency distribution of terms is influenced by various factors, which are not accounted for in the ideal Zipfian distribution.

### Q4 Removing stop words

The removal of stop words from the text corpus can impact the empirical distribution and the comparison with the Zipfian distribution in several ways. To evaluate the effect of stop word removal on the empirical and Zipfian distributions, I will compare the log-difference plots, the slopes of the empirical distributions, and the goodness-of-fit measures before and after stop word removal

```python
nltk.download('stopwords')

def count_terms_remove(filename):
    term_counts = {}
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))
    with open(filename, 'r', encoding='utf-8') as f:
        for line in f:
            tokens = word_tokenize(line.lower())
            for token in tokens:
                lemma = lemmatizer.lemmatize(token)
                if lemma.isalpha() and lemma not in stop_words:
                    term_counts[lemma] = term_counts.get(lemma, 0) + 1
    return term_counts
    
term_counts_remove = count_terms_remove('passage-collection.txt')

plot_zipf(term_counts_remove)
Diff(term_counts_remove)
```
